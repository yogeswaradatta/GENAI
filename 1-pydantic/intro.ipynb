pydantic basics creatin and using models

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str
    
person= Person(name='Krish',age=35,city='Banglore')
print(person)
```

```python
from typing import Optional
class Employee(BaseModel):
     id:int
     name:str
     department:str
     salary: Optional[float]=None
     is_active:Optional[bool]=True
     
employee1= Employee(name='Krish',id=35,department='Banglore')
print(employee1)
```

```python
from typing import List
from pydantic import BaseModel

class Classroom(BaseModel):
    room_number:str
    students:List[str]
    capacity:int
    
classroom1 = Classroom(room_number='2',students={'yogi','teja','surya'},capacity=3)
print(classroom1)
```
