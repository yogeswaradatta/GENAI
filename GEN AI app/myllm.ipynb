```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT1"]=os.getenv("LANGCHAIN_PROJECT1")
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader= WebBaseLoader("https://en.wikipedia.org/wiki/Surya")
loader
```

```python
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
documents=text.split_documents(docs)
documents
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings= OllamaEmbeddings(model="gemma:2b")
```

```python
from langchain_community.vectorstores import FAISS
vectorstoredb=FAISS.from_documents(documents,embeddings)

```
